```python
#https://bab2min.github.io/tomotopy/v0.12.2/en/
import tomotopy as tp
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import random
import nltk
nltk.download('stopwords')
nltk.download('punkt')
import pandas as pd
```

```python
sw = set(stopwords.words('english'))
```

```python
def tokenizer(doc,sw):
  word_list = []
  doc_strip = doc.strip()
  doc_lower = word_tokenize(doc_strip.lower())
  for word in doc_lower:
    if word not in sw and len(word) > 2:
      word_list.append(word)
  return(word_list)
```

```python
docs=[] # list of docs
f=open('hotelreviews.csv') # open the file
for doc in f: # for each doc in the file
    df = doc.partition('\t')
    docs.append(tokenizer(df[0],sw)) # tokenize the doc and add it to the list
f.close()
```

```python
docs
```

```python

#new LDA model
model = tp.LDAModel(k=6)


for doc in docs:# add the docs to the model
    model.add_doc(doc)


#train LDA model
for i in range(0, 500,10):
    model.train(10)
    print('Iteration: {}\tLog-likelihood: {}'.format(i, model.ll_per_word))
```

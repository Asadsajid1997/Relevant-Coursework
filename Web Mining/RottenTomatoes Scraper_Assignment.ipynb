```python

def scrape_query(driver, query:str,
                delay:int = 5 # number of seconds to wait 
                ):
    
    driver = driver
    #create a new csv writer for the story links
    fw=open('output.csv','w')
    writer = csv.writer(fw, lineterminator='\n')  # create a csv writer for this file
    writer.writerow(['Reviewer Name','Review_Polarity','Review Text', 'Review Date'])

    url=query # create the url

    driver.get(url) # visit the page

    input() # wait until we can click on the cookie overlay

    page_cnt=1 # keep track of page count

    while True: # keep going until there are no more pages

        print('page',page_cnt) # print current page count

        page_cnt+=1 # increment 
        
        #extract and write the links from the current page
        writer:csv.writer
            
    
    #get all story divs in the current page
        results=driver.find_elements(by=By.CSS_SELECTOR,value='div[class="review-row"]')
    
        for result in results: # for each story div
        
            name, polarity, text, date='NA','NA','NA', 'NA'
       
            try: # try to get the name and popularity
                name_box=result.find_element(by=By.CSS_SELECTOR,value='a[class="display-name"]')
                name=name_box.text
                print(name)
            except NoSuchElementException as e: # headline or link could not be found
                print('could not extract name')
            
            try:
                polarity_box=result.find_element(by=By.CSS_SELECTOR,value='score-icon-critic')
                polarity_state=polarity_box.get_attribute('state')
                if 'fresh' in polarity_state:
                    polarity = 'fresh'
                elif 'rotten' in polarity_state:
                    polarity = 'rotten'
                print(polarity)

            
            except NoSuchElementException as e: # headline or link could not be found
                print('could not extract polarity')
    
            try: # try to get the text
                text_box=result.find_element(by=By.CSS_SELECTOR,value='p[class="review-text"]')
                text=text_box.text
                print(text)
            except NoSuchElementException as e: # headline or link could not be found
                print('could not extract text')
            
            try: # try to get the date
            
            #get the date box
                date_box=result.find_element(by=By.CSS_SELECTOR,value='span[data-qa="review-date"]')

            #get the date string
                date=date_box.text
                print(date)
            except NoSuchElementException as e:
                print('could not extract date')
            
        #write a new row for this story
            writer.writerow([name, polarity, text, date])
        
        try:
        #wait until the next button appears                                                                        
            next_button = WebDriverWait(driver, delay).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'rt-button[class*="next"]')))

        #check if the next button is clickable
        except TimeoutException:
            print('no more pages')
            break

        #click the next button and wait 
        #next_button.click()
        driver.execute_script("arguments[0].click();", next_button)
        time.sleep(delay)
  
    fw.close()
```

```python
scrape_query(webdriver.Chrome(service=Service(ChromeDriverManager().install())),
             'https://www.rottentomatoes.com/m/exodus_gods_and_kings/reviews')
```
